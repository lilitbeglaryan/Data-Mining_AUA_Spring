```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import seed
seed(1914)
```

```python
x1 = np.random.uniform(low=0.0, high=13.0, size=30000)
y1 = np.random.uniform(low=0.5, high=3.0, size=30000)
c1 = np.full(shape=30000, fill_value=0, dtype=int)

x2 = np.random.uniform(low=0.0, high=13.0, size=20000)
y2 = np.random.uniform(low=3.2, high=5.0, size=20000)
c2 = np.full(shape=20000, fill_value=1, dtype=int)

x3 = np.random.uniform(low=0.0, high=13.0, size=40000)
y3 = np.random.uniform(low=5.2, high=10.0, size=40000)
c3 = np.full(shape=40000, fill_value=2, dtype=int)

x4 = np.random.uniform(low=-0.5, high=13.5, size=40000)
y4 = np.random.uniform(low=0.0, high=10.5, size=40000)
c4 = np.full(shape=40000, fill_value=3, dtype=int)

x = np.concatenate((x4, x2, x3, x1), axis=None)
y = np.concatenate((y4, y2, y3, y1), axis=None)
c = np.concatenate((c4, c2, c3, c1), axis=None)
```

```python
df = pd.DataFrame({'x' : x, 'y' : y, 'c' : c})
```

```python
plt.close()
plt.figure(figsize=(10, 13))
sns.scatterplot(data = df, x='x', y='y', hue='c', legend=False, palette=["#1c5896", "#b52644", "#c5270d", "#c35150"])

plt.show()
```
