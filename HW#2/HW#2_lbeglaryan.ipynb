```python
def plot_boxplot(var_name, dataset = df):
    plt.figure(figsize = (8, 6))
    sn.boxplot(data = dataset, x = "Class/ASD", y = var_name)
    plt.title(f'Distribution of variable "{var_name}"')
    
def plot_objects(var_name,dataset = df):
    sn.countplot(x = dataset[var_name].astype("category"))
    plt.title(f'Distribution of variable "{var_name}"')
```
